### Sentiment Analysis
using VADER and TextBlob

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
sid_obj= SentimentIntensityAnalyzer()
print(sid_obj.polarity_scores("#Covid_19SouthAfrica\nThis is very serious")) 
```

```python
list(sid_obj.polarity_scores("SouthAfrica: No COVID-19 vaccine yet").values())[1]
```

```python
tweet_df['Negative']="" 
tweet_df['Neutral']="" 
tweet_df['Positive']="" 
tweet_df['Sentiment_VADER']= ""
```

```python
def get_senti(text):
    sentiment = list(sid_obj.polarity_scores(text).values())[3]
    if sentiment > 0:
        sentiment_label = 'Positive'
    elif sentiment < 0:
        sentiment_label = 'Negative'
    else:
        sentiment_label = 'Neutral'
    result = sentiment_label
    return result
```

```python
get_senti("South Africa. New variant, C.1.2, may be more ")
```

```python
count = 0

for row in tweet_df['cleaned']:
    
    tweet_df['Negative'][count]= list(sid_obj.polarity_scores(row).values())[0]
    tweet_df['Neutral'][count]= list(sid_obj.polarity_scores(row).values())[1]
    tweet_df['Positive'][count]=list(sid_obj.polarity_scores(row).values())[2]
    count += 1
senti = lambda x : get_senti(x)
tweet_df['Sentiment_VADER'] = tweet_df['cleaned'].apply(senti)
```

```python
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

stopwords = set(STOPWORDS)
```

```python
df.head()
#Checking for NaN values
df.isna().sum()
#Removing NaN Values
df.dropna(inplace = True)
#Creating the text variable
text = " ".join(cat.split()[1] for cat in df['clean text'])
```

```python
stop_words = ["covid19", "says", "said", "covid","africa","african","south"] + list(STOPWORDS)

wordcloud = WordCloud(width = 5000, height = 2000, random_state=1, 
                      background_color='white', colormap=sns.color_palette("Blues", as_cmap=True), collocations=False, stopwords = stop_words
                      ).generate(text)



# create coloring from image
plt.figure(figsize=[120,80])
plt.imshow(wordcloud)
plt.axis("off")
# store to file
#plt.savefig("news.png", format="png") 
plt.show()
```
